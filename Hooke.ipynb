```python
import numpy as np

import matplotlib.pyplot as plt

#teorik alt yapı için buraya bakmıştım: https://www3.nd.edu/~apilking/Math10560/Lectures/wsheetHmotion%20copy/worksheet_Harmonic_MotionSolutions%20copy.pdf


def Hooke(x0,v0,k,m):
    x = x0

    v = v0

    a0 = -k*x0/m

    a = a0

    t = 0

    dt = 0.01

    vdata = [] #hala daha numpy arrayleri ile beceremiyorum

    xdata = []

    adata = []

    tdata = []

    

    while t < 20:

        vdata.append(v)

        xdata.append(x)

        adata.append(a)

        tdata.append(t)

        t = t + dt

        a = -k*x/m

        v = v + a*dt

        x = x + v*dt

    print(x, v, a)#!aşağıda yazdırdığım değerlere bakarsak ivmenin bu kadar çıkması normal mi?

    plt.figure(figsize = (15, 15))
    plt.plot(tdata, xdata, label='Yaklaşık')
    plt.title('Euler Methodu Hooke Yasası ')
    plt.xlabel('t')
    plt.ylabel('x')
    plt.grid()
    plt.legend(loc='upper right')
    plt.show()

    #! Teoride bir yerlerde hata olduğunu seziyorum ama göremiyorum!
```

```python
Hooke(11, 75, 11, 1)
```
